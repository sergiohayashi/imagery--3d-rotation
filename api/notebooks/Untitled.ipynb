```python
data = [47.50,
47.50,
50.00,
57.50,
57.50,
45.00,
47.50,
47.50,
62.50,
57.50,
55.00,
57.00,
55.00,
62.50,
50.00,
45.00,
50.00,
52.00,
]
```

```python
import statistics
```

```python
# Median
median = statistics.median(data)

# Variance (sample variance)
variance = statistics.variance(data)

# Standard deviation (sample)
std_dev = statistics.stdev(data)

print("Median:", median)
print("Variance:", variance)
print("Standard deviation:", std_dev)
```
